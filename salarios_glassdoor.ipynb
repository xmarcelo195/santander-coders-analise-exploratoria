### Salários Glassdoor

**Objetivo**: Análise

```python
df = pd.read_csv('https://raw.githubusercontent.com/xmarcelo195/santander-curso-3/main/glassdoor_jobs.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.tail(10)
```

```python
df.columns
```

```python
## Tratamentos
```

```python
#salarios
#create min salary
df['min_salary'] = df['Salary Estimate'].apply(lambda x: str(x).split("K")[0].replace("$",""))


def get_max_salary(x:str) -> str:
    result = str(x).split("(")[0].split("-")[-1].replace("$","").replace("K","")
    return result

#create max_salary
df['max_salary'] = df['Salary Estimate'].apply(lambda x: get_max_salary(x))

#remove hour_salaries
df = df[~df['min_salary'].str.contains('Hour')]

def provided_salary(x: str) -> str:
    x = str(x)
    if 'Employer Provided Salary' in x:
        result = x.split(':')[-1]
    else:
        result = x
    return result

#tratar provided salary:
df['min_salary'] = df['min_salary'].apply(lambda x: provided_salary(x))

#remove empty salaries
df = df[df['min_salary'] != '-1']
df = df[df['max_salary'] != '1']


# average salary
df['avg_salary'] = (df['min_salary'].astype(int) + df['max_salary'].astype(int))/2
```

```python
df['python'] = np.where(df['Job Description'].str.lower().str.contains('python'), 1, 0)
df['aws'] = np.where(df['Job Description'].str.lower().str.contains('aws'), 1, 0)
df['spark'] = np.where(df['Job Description'].str.lower().str.contains('spark'), 1, 0)
df['excel'] = np.where(df['Job Description'].str.lower().str.contains('excel'), 1, 0)
df['azure'] = np.where(df['Job Description'].str.lower().str.contains('azure'), 1, 0)
df['cloud'] = np.where(df['Job Description'].str.lower().str.contains('cloud'), 1, 0)
```

```python
#observar como vem a lista de competidores
df['Competitors'].unique()
```

```python
#count competitors
def count_competitors(x: str)->str:
    x = str(x)
    if x == "-1":
        return 0
    else:
        result = len(x.split(','))
    return result

df['n_comp'] = df['Competitors'].apply(lambda x: count_competitors(x))
```

```python
df = df[['Job Title','min_salary', 'max_salary', 'avg_salary','n_comp','python','aws','spark','excel','azure','cloud']]
df.head()
```
